```python
def evaluate(tensors):
    """Evaluates Tensor or EagerTensor to Numpy `ndarray`s.
    Args:
    tensors: Object of `Tensor` or EagerTensor`s; can be `list`, `tuple`,
        `namedtuple` or combinations thereof.

    Returns:
        ndarrays: Object with same structure as `tensors` except with `Tensor` or
          `EagerTensor`s replaced by Numpy `ndarray`s.
    """
    if tf.executing_eagerly():
        return tf.contrib.framework.nest.pack_sequence_as(
            tensors,
            [t.numpy() if tf.contrib.framework.is_tensor(t) else t
             for t in tf.contrib.framework.nest.flatten(tensors)])
    return sess.run(tensors)
```

```python

def session_options(enable_gpu_ram_resizing=True, enable_xla=True):
    """
    Allowing the notebook to make use of GPUs if they're available.
    
    XLA (Accelerated Linear Algebra) is a domain-specific compiler for linear 
    algebra that optimizes TensorFlow computations.
    """
    config = tf.ConfigProto()
    config.log_device_placement = True
    if enable_gpu_ram_resizing:
        # `allow_growth=True` makes it possible to connect multiple colabs to your
        # GPU. Otherwise the colab malloc's all GPU ram.
        config.gpu_options.allow_growth = True
    if enable_xla:
        # Enable on XLA. https://www.tensorflow.org/performance/xla/.
        config.graph_options.optimizer_options.global_jit_level = (
            tf.OptimizerOptions.ON_1)
    return config


def reset_sess(config=None):
    """
    Convenience function to create the TF graph & session or reset them.
    """
    if config is None:
        config = session_options()
    global sess
    tf.reset_default_graph()
    try:
        sess.close()
    except:
        pass
    sess = tf.InteractiveSession(config=config)

```

```python
# Initialize a single 3-dimension vMF distribution.
mu = [[0., 1, 0]]
conc = [[20.]]
vmf = tfd.VonMisesFisher(mean_direction=mu, concentration=conc)
uniform = tfd.VonMisesFisher(mean_direction=mu, concentration=0.)
```

```python
sample_size = 1000
sampled_values = vmf.sample()
sampled_values
```

We may compute KL divergence between von-Mises-Fisher distribution on 3D sphere and uniform distribution thereon.

```python
def _kl_divergence_uniform(mu_q, kappa_q, d):
    "Assuming d is odd and "
    d_star = float(d)/2 - 1
    d_diamond = (float(d) - 3)/2
    d_circle = (float(d) - 1)/2
    
    return kappa_q - d_star*np.log(2)
    
```

```python
[sampled_values_] = evaluate([sampled_values])
sampled_values_.shape
```
